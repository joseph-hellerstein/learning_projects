# Vector basics

## Definitions

1. Multiplication of the scalars $x, y$ is denoted by $x \cdot y$.
1. A point is a linear combination of the vector coordinates $i, j, k$
1. A (direction) vector has the point $<0, 0, 0>$ as its origin, and a point as its endpoint.
1. The length of a direction vector $p = <p_1, p_2, p_3>$ is $\sqrt{\sum_i p_i^2}$, and is denoted by $||p||$.
1. A line has a direction vector $v$ and a point $p$ and includes the points $q = kv + p$ for $k \in (-\infty, \infty)$.
1. The cross product of the vectors $p, q$ is $p \circ q = \sum_i p_i \cdot q_i$.
1. A **plane** is defined by a point and a normal to the plane. Let $p$ be a point and $v$ be the normal vector. The plane consists of all points such that
\begin{eqnarray}
q \backepsilon \sum_i v_i (q_i - p_i) & = & 0
\end{eqnarray}
1. The crossproduct between two vectors is a third vector that is orthogonal to the first two.

## Results

1. The cosine of the angle between the vectors $p$, $q$ is $\frac{p \circ q}{||p|| \cdot ||q||}$.
1. The **projection** of the vector $p$ on the vector $q$ is base of the right triangle with $k \cdot p$ as the base and $q$ as the hypotenuse. Let $\theta$ be the angle between $p, q$, and so $cos \theta = \frac{p \circ q}{||p|| \cdot ||q||}$. So, the length $L$ is
\begin{eqnarray}
L & = & cos \theta \cdot ||p|| \cdot \frac{q}{||q||} \\
& = & \frac{p \circ q}{||q||^2} q
\end{eqnarray}
1. Given $P_i = (p_i, v_i)$ for points $p_i$ and vectors $v_i$, the angle between the planes is the angle between $v_1, v_2$. The planes are non-intersecting if $v_1$ is a multiple of $v_2$.

### Cross product

Let $p, q$ be vectors. Their cross product is denoted by $p \times q$. It is defined as
$\left|
\begin{pmatrix}
i & j & k \\
p_1 & p_2 & p_3 \\
q_1 & q_2 & q_3
\end{pmatrix}
\right|$, where vertical bars indicate determinant.

Proof that the result is a vector that is orthogonal to $p, q$.
1. The result is a vector that is the sum of $i, j, k$. That is, from definition of the determinant,
$
r =
\begin{pmatrix}
p_2 & p_3 \\
q_2 & q_3 \\
\end{pmatrix}
i -
\begin{pmatrix}
p_1 & p_3 \\
q_1 & q_3 \\
\end{pmatrix}
j +
\begin{pmatrix}
p_1 & p_2 \\
q_1 & q_2 \\
\end{pmatrix}
k,
$
which is clearly a vector since it is a linear combination of vectors.
1. Now consider the dot product $r \circ p$. We show that this is 0.
$r \circ p =
\begin{pmatrix}
p_2 & p_3 \\
q_2 & q_3 \\
\end{pmatrix}
p_1 -
\begin{pmatrix}
p_1 & p_3 \\
q_1 & q_3 \\
\end{pmatrix}
p_2 +
\begin{pmatrix}
p_1 & p_2 \\
q_1 & q_2 \\
\end{pmatrix}
p_3
$
But this is equivalent to
$\left|
\begin{pmatrix}
p_1  & p_2 & p_3 \\
p_1 & p_2 & p_3 \\
q_1 & q_2 & q_3
\end{pmatrix}
\right|
,$ which is 0. Clearly, this is true for $q$ as well. Q.E.D.

# Review of Conic Sections

## Elipse

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$

```python
sp.plotting.plot_implicit(sp.Eq(x**2/4 + y**2, 3), (x, -10, 10), (y, -3, 3))
```

## Parabola

$$y = ax^2$$

```python
sp.plotting.plot_implicit(sp.Eq(y - x**2, 3), (x, -3, 3), (y, -10, 10))
```

## Hyperbola

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$$

```python
sp.plotting.plot_implicit(sp.Eq(x**2/4 - y**2, 3), (x, -10, 10), (y, -3, 3))
```

# Cylinders and Quadric Surfaces

## Basics

A quadratic surface has the form $P(x, y, z) = 0$, where $P$ is a second order polynomial in $x, y, z$.

We can classify quadratic services based on their traces, the type of conic section that occurs when intersected with the $xy, yz, xz$ planes.

A **cylinder** is any surface that is formed by "extending" a plane. For example, a circle in the $xy$ plane can be extended by letting $z$ take an arbitrary value.

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$$

## Elispoids

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2}= 1$$

```python
expr = sp.sqrt(1 - x**2 - y**2)
plot3d(expr, -expr, (x, -1, 1), (y, -1, 1), show=True)
```

## Paraboloid of One Sheet

This is a hyperboloid with one section because, if it is centered at $<0, 0, 0>$, the origin is contained within the surface.

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} + \frac{z^2}{c^2}= 1$$

```python
expr =  sp.sqrt(1 - x**2/2 + y**2)
plot3d(expr, -expr,  (x, -10,10), (y, -10,10), show=True)
```

## Paraboloid of Two Sheets

This is a hyperboloid with two sections to it because, if it is centered at $<0, 0, 0>$, the origin is not contained within the surface.

$$-\frac{x^2}{a^2} - \frac{y^2}{b^2} + \frac{z^2}{c^2}= 1$$

```python
expr =  sp.sqrt(1 + x**2/2 + y**2)
plot3d(expr, -expr,  (x, -10,10), (y, -10,10))
```

## Hyperbaloid

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} + z = 1$$

```python
expr =  1 - x**2/2 - y**2
plot3d(expr,  (x, -10,10), (y, -10,10))
```

## Saddle

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} + z = 1$$

```python
expr =  1 - x**2/2 + y**2
plot3d(expr,  (x, -10,10), (y, -10,10))
```
